```python
# Importando os pacotes a serem utilizados
import pandas as pd
import os
import warnings
warnings.filterwarnings("ignore")
```

```python
# Dica Bonus - Leitura de um arquivo dispon√≠vel na Web
df = pd.read_excel('https://docs.google.com/uc?export=download&id=1SfyiP2z2HJNd7iMGx1Mt82WbACkJDm7I')
```

```python
# Carregando diversos arquivos simultaneamente
lista_dataFrames = []
lista_nome = []
diretorio = "D:\\Testes"
for arq in os.listdir(diretorio):
    if arq.endswith(".csv"):  
        nome_arquivo = "\\" + arq        
        nome_df = pd.read_csv(diretorio+nome_arquivo, sep=';' , encoding='latin-1')        
        print("Nome do arquivo: ", arq)
        lista_dataFrames.append(nome_df)         
        #x = arq.rfind(".")
        #lista_nome.append(arq[:x])
        #print(arq[:x])        
```

```python
# Concatenando os arquivos em um √∫nico DataFrame
contador = 1
df_concatenado = lista_dataFrames[0]
for i in lista_dataFrames[1:]:
    df_concatenado = pd.concat([df_concatenado, lista_dataFrames[contador]])
    contador += 1    
```
