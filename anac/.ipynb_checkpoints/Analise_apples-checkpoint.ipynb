```python
import numpy as np
```

```python
# Carregar o arquivo CSV usando loadtxt
dados = np.loadtxt('apples_ts.csv', delimiter=',', dtype=str)
dados
```

```python
dados.shape
```

```python
for linha in dados:
    for i, valor in enumerate(linha):
        if not valor.strip():
            linha[i] = '0'
dados
```

```python
for i in range(6):
    dados[i][0] = i
dados
```

```python
dados = dados.astype(float)
```

```python
dados
```

```python
dados[np.isnan(dados)] = 0
```

```python
sub_array = dados[:, 0:13]
sub_array
```

```python
for i in dados[0, :13]:
    sub_array = i
sub_array
```

```python
sub_array = dados[1:, 1:13]
sub_array
```

```python
sub_array.shape
```

```python
# 1. Tendência geral de preços ao longo do tempo
media_por_mes = np.mean(sub_array, axis=1)
print("\nTendência geral de preços ao longo do tempo:")
print(media_por_mes)
```
